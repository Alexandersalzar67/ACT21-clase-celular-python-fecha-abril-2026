```python
class Celular:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        self.encendido = False

    def encender(self):
        self.encendido = True
        print(f"El {self.modelo} se ha encendido.")
+
# Uso de la clase
celular1 = Celular("Apple", "iPhone 15")
print(f"Celular: {celular1.marca} {celular1.modelo}, Precio: $1000")
celular1.encender()

celular2 = Celular("Samsung", "Galaxy S23")
print(f"Celular: {celular2.marca} {celular2.modelo}, Precio: $800")
celular2.encender()

print(" creador de software Alexander Salazar NL = 49")
```
